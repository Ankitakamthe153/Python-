# Sales Analysis Case Studies

```python
# Task1: import pandas
import pandas as pd

# Task 2: Read CSV files and create Data Frames
df = pd.read_csv("Sale.csv")

# Task 3: Perform the operation

df.head()  # head() gives first specific rows by default
```

```python
df.head(10) # head() gives first five rows by default
```

```python
df.tail() # tail gives last 5 coloumn
```

# Augment with additional coloumn

```python
# add month 
df['Month']=df['Order Date'].str[0:2]
df.head()
```

### Cleaning up the Data becouse NaN value need to remove


```python
#find the NaN
nan_df= df[df.isna().any(axis=1)]
nan_df.head()

#drop NaN
clean_df=df.dropna(how='all')
clean_df.head()


```

```python
#Find or
or_df= clean_df[clean_df['Order Date'].str[0:2]=='Or']
or_df.head()
```

```python
#Take the data without Or
clean_df= clean_df[clean_df['Order Date'].str[0:2]!='Or']
clean_df.head()
```

```python
#Convert into specific Data type
clean_df['Month']=clean_df['Order Date'].str[0:2]
clean_df['Month']=clean_df['Month'].astype('bool')
clean_df.head()
```

```python
#Convert into specific Data type

clean_df['Quantity Ordered']=pd.to_numeric(clean_df['Quantity Ordered'])
clean_df['Price Each']=pd.to_numeric(clean_df['Price Each'])
clean_df.head()
```

```python

#multipy the Quantity and Price

clean_df['Sales']=clean_df['Quantity Ordered']*clean_df['Price Each']
clean_df.head()
```

```python
# Question: What was the best month for sales and how much earned in that month

results=clean_df.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt

months=range(1,13)
plt.bar(months, results['Sales'])

plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month Numbers')

plt.show()


```

```python
#which city having most sale

def get_city(address):
    return address.split(',')[1]

clean_df['city']=clean_df['Purchase Address'].apply(lambda x:get_city(x))


def get_state(address):
    return address.split(',')[2]

clean_df['state']=clean_df['Purchase Address'].apply(lambda y:get_state(y))
clean_df.head()
```

```python
# Question 2: What was the best city for sales and how much earned in that city

results=clean_df.groupby('city').sum()
results
```

```python
import matplotlib.pyplot as plt

#cities=clean_df['city'].unique()

cities=[city for city, df in clean_df.groupby('city')]

plt.bar(cities, results['Sales'])

plt.xticks(cities,rotation='vertical', size=8)
plt.yticks(results['Sales'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')

plt.show()
```

```python
import matplotlib.pyplot as plt

#cities=clean_df['city'].unique()

cities=[city for city, df in clean_df.groupby('city')]

plt.scatter(cities, results['Sales'])

plt.xticks(cities,rotation='vertical', size=8)
plt.yticks(results['Sales'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')

plt.show()
```

```python
# Question 3: What was the best Sate for sales and how much earned in that state

results=clean_df.groupby('state').sum()
results
```

```python
import matplotlib.pyplot as plt

#cities=clean_df['city'].unique()

states=[state for state, df in clean_df.groupby('state')]

plt.bar(states, results['Sales'])

plt.xticks(states,rotation='vertical', size=8)
plt.yticks(results['Sales'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('State Name')

plt.show()
```

```python
import matplotlib.pyplot as plt

#cities=clean_df['city'].unique()

states=[state for state, df in clean_df.groupby('state')]

plt.scatter(states, results['Sales'])

plt.xticks(states,rotation='vertical', size=8)
plt.yticks(results['Sales'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('State Name')

plt.show()
```

```python
# What time should we display advertisement to maxmise likelyhood customer that buying product

import pandas as pd
clean_df['Order Date']=pd.to_datetime(clean_df['Order Date'])
clean_df['Hour']=clean_df['Order Date'].dt.hour
clean_df['Minute']=clean_df['Order Date'].dt.minute
clean_df.head()
```

```python
results=clean_df.groupby(['Hour']).count()
results.head()
```

```python
import matplotlib.pyplot as plt

#cities=clean_df['city'].unique()

Hours=[Hour for Hour, df in clean_df.groupby('Hour')]
plt.plot(Hours, clean_df.groupby(['Hour']).count(), 'o--r',ms='10', mec="blue", mfc="yellow")

plt.xticks(Hours)
plt.title('likelyhood customer')
plt.ylabel('Hourly Total Customer')
plt.xlabel('Hour')
plt.grid()
plt.show()
```

```python
#Sales per Hour

import matplotlib.pyplot as plt

#cities=clean_df['city'].unique()

Hours=[Hour for Hour, df in clean_df.groupby('Hour')]

plt.bar(Hours, results['Sales'])
plt.xticks(Hours)
plt.title('Sales per Hour')
plt.ylabel('Hourly Total Customer')
plt.xlabel('Hour')
plt.show()
```
